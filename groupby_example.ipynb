```python
import pandas as pd
```

```python
cheers=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
cheers
```

```python
cheers.groupby('continent').beer_servings.mean()
```

```python
cheers.groupby('continent').wine_servings.describe()
```

```python
cheers.groupby('continent').mean()
```

```python
cheers.groupby('continent').median()
```

```python
cheers.groupby('continent').spirit_servings.agg(['mean','max','min'])
```
